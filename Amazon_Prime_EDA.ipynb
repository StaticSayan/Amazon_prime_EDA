#### Movies vs TV Shows in Each Genre (Stacked Bar) 

```python
# visualization code
# Expand genres into multiple rows for accurate crosstab
data_expanded = data.explode('genres')
import ast
data['genres'] = data['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else [x])
data_expanded = data.explode('genres')
genre_type = pd.crosstab(data_expanded['genres'], data_expanded['type'])
top_10_genres = genre_type.sum(axis=1).sort_values(ascending=False).head(10).index
genre_type = genre_type.loc[top_10_genres]
plt.figure(figsize=(12, 6))
genre_type.plot(kind='bar', stacked=True, colormap='viridis')
plt.title("Genre vs Content type (Top 10 Genres)")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

**Stacked bar** charts illustrate the relationship between two categorical variables. This helps explore how different genres split
between movies and shows, guiding content planning.

#### Average IMDb Rating Over Years by Content Type:

```python
# visualization code
# Clean and prepare data
data['type'] = data['type'].str.upper()
data['release_year'] = pd.to_numeric(data['release_year'], errors='coerce')
data['imdb_score'] = pd.to_numeric(data['imdb_score'], errors='coerce')
# Filter out missing values
filtered = data.dropna(subset=['release_year', 'type', 'imdb_score'])
# Group and calculate average IMDb score
avg_ratings = filtered.groupby(['release_year', 'type'])['imdb_score'].mean().unstack()
#  Plot
plt.figure(figsize=(14, 6))
avg_ratings.plot(marker='o', linewidth=1)
plt.title("Average IMDb Rating Over Years by Content Type", fontsize=14)
plt.xlabel("Release Year")
plt.ylabel("Average IMDb Score")
plt.legend(title="Content Type")
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

The chart above directly addresses content quality trends over time and compare whether TV shows or movies have improved or declined in viewer ratings.
Valuable for stakeholders analyzing platform quality strategy over time.

##### 2. What is/are the insight(s) found from the chart?

Over the years IMDb ratings for movies run around 5.5 - 6.5/7 more or less, where as compared to this TV shows have a higher rating 
level in genreal(though there are some lower ratings).

##### 3. Will the gained insights help creating a positive business impact?

### Pair Plot of IMDb Score, Runtime & Seasons

```python
# Pair Plot visualization code
# Filter dataset to drop rows with missing essential values
pair_data = data[['imdb_score', 'runtime', 'seasons']].dropna()
# Create a pair plot
sns.pairplot(pair_data, diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle("Pair Plot of IMDb Score, Runtime & Seasons", y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

#### Reason to choose this chart :

To understand how quality (IMDb score) and content length (runtime/seasons) vary over time and whether certain patterns emerge that can guide content production, licensing decisions, or user engagement strategies.

##### 2. What is/are the insight(s) found from the chart?

The pair plot reveals the following key insights:
* IMDb Score Distribution: Most titles have IMDb scores between 5 and 8, indicating average to moderately high audience ratings.
* Runtime vs. Seasons: Titles with higher runtimes tend to have fewer seasons, suggesting longer content is typically standalone (e.g., movies), while    multi-season shows usually have shorter episode runtimes.
* Seasons Distribution: A large number of titles have only 1–2 seasons, indicating a dominance of limited series or short-format content.
* No Strong Correlation: There is no strong linear correlation between IMDb scores and runtime or number of seasons, indicating that audience rating is   relatively independent of content length or series duration.

### 3.Will the gained insights help creating a positive business impact?

These insights can positively influence content strategy:
* Focus on highly rated short-format content or limited series to align with existing user preferences.
* Invest in quality over quantity for series to improve IMDb ratings, as longer runtimes or multiple seasons don't guarantee better reception.
* Improve recommendations and marketing by segmenting content based on runtime and season count preferences.
